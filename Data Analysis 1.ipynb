### Inicialmente serão analisadas possíveis inconsistências no dataset. Limpeza de informações nulas, duplicadas e informações que não possuem sentido.


```python
# Analisando se há alguma célula vazia no dataset
roof.info()
```

```python
# Arrumando a data
# 1. Retirando a string T00... no final das células
data = roof['data_venda'].str[:8]
roof['data_venda'] = data


# 2. Formatando a data para aparecer na ordem que eu quero
roof['data_venda']  = pd.to_datetime(roof['data_venda'], format='%Y-%m-%d')

# 3. Criando colunas para mês, dia e ano (para possibilitar análises futuras)
roof['dia_venda'] = pd.DatetimeIndex(roof['data_venda']).day
roof['mes_venda'] = pd.DatetimeIndex(roof['data_venda']).month
roof['ano_venda'] = pd.DatetimeIndex(roof['data_venda']).year
```

```python
roof
```

```python
# Encontrando a cidade de cada imóvel
!pip install uszipcode
from uszipcode import SearchEngine, SimpleZipcode

search = SearchEngine()
def zco(x):
    city = search.by_zipcode(x).major_city
    return city if city else 'None'

roof['city'] = roof['zipcode'].apply(zco)
```

```python
# Acrescentando coluna de valor de cada pés quadrado
roof['preco_area'] = roof['preco_venda']/roof['area_terreno']
```

```python
# Acrescentando coluna de aproveitamento do terreno (tamanho do terreno dividido pela area habitável)
roof['aproveitamento_terreno'] = roof['area_habitavel']/roof['area_terreno']
```

```python
roof
```

```python
# Analisando o tipo de variável de todas as colunas
roof.info()
```

```python
# Adequando os tipos de dados de cada coluna
roof['qtd_banheiros'] = roof.qtd_banheiros.astype('int64')
roof['qtd_andares'] = roof.qtd_andares.astype('int64')
roof['area_habitavel'] = roof.area_habitavel.astype('float64')
roof['area_terreno'] = roof.area_terreno.astype('float64')
roof['area_terrea'] = roof.area_terrea.astype('float64')
roof['area_subsolo'] = roof.area_subsolo.astype('float64')
roof['area_habitavel_15vizinhos'] = roof.area_habitavel_15vizinhos.astype('float64')
roof['area_terreno_15vizinhos'] = roof.area_terreno_15vizinhos.astype('float64')
roof['city'] = roof.city.astype('category')
```

```python
# Para analisar os preços das casas por cidade para identificar as cidades mais caras, decidi usar um agrupamento por média
roof_grouped = roof.groupby(['city']).mean().sort_values('preco_venda', ascending= False).reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(x= roof_grouped['city'], y= roof_grouped['preco_venda'], marker_color= 'blue'))
fig.update_layout(title='Mediana dos preços por cidade', title_x = 0.5)
```

```python
# Com a análise, percebi que a cidade de Mercer Island estava fugindo dos padrões de preço e decidi retirar as casas dessa cidade da relação
mercer = roof.loc[roof['city'] == 'Mercer Island']
roof.drop(mercer.index, inplace= True)
roof
```

```python
# Criando um loop para salvar em uma lista todos os IDs que foram vendidos mais de uma vez
todos= []
repetidos = []
for i in roof['ID']:
  if i in todos:
    repetidos.append(i)
  else:
    todos.append(i)
```

```python
# Conferindo as quantidades de repetidos
print(roof['ID'].duplicated().sum(),'=',len(repetidos))
```

```python
# Certificando se houve algum imóvel vendido mais de 2 vezes para decidir a estratégia para usar
rep = roof[roof['ID'].isin(repetidos)].sort_values(by=['ID'])
rep.stb.freq(['ID'])
```

```python
roof[roof['ID'] == 2619920170]
```

```python
# Criando um dataset agrupado por ID e puxando apenas a venda mais recente do ID
venda1 = roof.groupby(['ID']).agg({'data_venda':'max'}).reset_index()
venda1
```

```python
# Criando um dataset agrupado por ID e puxando apenas a venda mais antiga do ID
venda2 = roof.groupby(['ID']).agg({'data_venda':'min'}).reset_index()
venda2
```

```python
# Com os dois datasets agrupados criados, vou usá-los para fazer a conta da diferença de preço de venda da venda mais recente para a mais antiga. Com esse resultado, saberei dizer se o
# o imóvel valorizou ou não e ter um insight sobre como é a expectativa de futuro para os preços daquela região.
```

```python
# Criando uma coluna preco venda para os dois datasets
venda1 = venda1.assign(preco_venda=0)
venda2 = venda2.assign(preco_venda=0)
```

```python
# Criando um loop para acrescentar o preço de venda da venda mais antiga
for z,(id,dv) in enumerate(zip(venda2['ID'],venda2['data_venda'])):
  x = roof['preco_venda'][(roof['ID'] == id) & (roof['data_venda'] == dv)]
  for i in x:
    venda2['preco_venda'][venda2.index == z] = i
```

```python
venda2
```

```python
# Criando um loop para acrescentar o preço de venda da venda mais recente
for z,(id,dv) in enumerate(zip(venda1['ID'],venda1['data_venda'])):
  x = roof['preco_venda'][(roof['ID'] == id) & (roof['data_venda'] == dv)]
  for i in x:
    venda1['preco_venda'][venda1.index == z] = i
```

```python
venda1
```

```python
# Criando uma coluna de valorização (ou desvalorização) no dataset venda1

venda1['valorization'] = venda1['preco_venda'] - venda2['preco_venda']
venda1[venda1['ID'].isin(repetidos)]
```

```python
# Criando uma coluna de valorização no dataset principal e acrescentando as valorizações das casas vendidas mais de uma vez
roof = roof.assign(valorization=0)
roof['valorization'][roof['ID'] == 2619920170] = -7500
roof['valorization'][roof['ID'] == 3395040550] = 70000
roof['valorization'][roof['ID'] == 7853400250] = 35000
roof['valorization'][roof['ID'] == 8648900110] = 0
```

```python
# Criando um dataset apenas com os imóveis que houveram revenda e valorização
val = roof[roof['valorization'] != 0]
```

```python
# Identificando as cidades com maior tendência de valorização e desvalorização, a partir do dataset criado anteriormente
plt.figure(figsize = (15,10))
plt.bar(val['city'],val['valorization'], color='g');
plt.xlabel('Cidades')
plt.ylabel('Valorização')
plt.title('Valorização por cidade')
plt.grid(visible=True, color = 'gray', linestyle= ':', linewidth = 0.4)
plt.axhline(y=0, xmin=0, xmax=1, color = 'red', **{'ls':'--', 'lw':2.5});
```
